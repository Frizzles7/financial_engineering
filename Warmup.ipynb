# Warmup

Initial warmup for financial engineering.  Verify can create some samples, graphs, and calculate basic metrics.

```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# generate 1000 samples from a normal(0,1) distribution
noise = np.random.normal(0, 1, 1000)
```

```python
# mean should be close to 0
np.mean(noise)
```

```python
# standard deviation should be close to 1
np.std(noise, ddof=1)
```

```python
# plot our sample as a time series
plt.plot(noise)
plt.show()
```

```python
# plot our sample as a histogram
plt.hist(noise)
plt.show()
```

```python
# add a trend line to the noise
t = np.arange(1000)
trend = 0.01*t + noise
```

```python
# create a scatter plot of the trended data
plt.scatter(t, trend)
plt.show()
```

```python
# use np.cumsum() on the noise
cumulative_noise = np.cumsum(noise)
```

```python
# plot the cumulative noise as a time series
plt.plot(cumulative_noise)
plt.show()
```

```python
# generate 1000 samples from a multivariate normal distribution
# means = [0, 0], cov matrix = [[1, -0.5],[-0.5, 2]]
means = [0, 0]
cov = [[1, -0.5],[-0.5, 2]]
x, y = np.random.multivariate_normal(means, cov, 1000).T
```

```python
# plot x and y as a scatter plot
plt.scatter(x, y)
plt.show()
```

```python
# verify the means are close to zero (w/o using built in functions)
avg_x = sum(x) / 1000
avg_y = sum(y) / 1000
print(avg_x, avg_y)
```

```python
# verify the sample covariance matrix is close to what we defined
cov_x_y = sum((x-avg_x)*(y-avg_y))/1000
var_x = sum((x-avg_x)**2)/1000
var_y = sum((y-avg_y)**2)/1000
print(cov_x_y, var_x, var_y)
```
