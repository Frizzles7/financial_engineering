# PACF and ACF

```python
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate iid noise from the std normal dist
x0 = np.random.randn(1000)
```

```python
plt.plot(x0);
```

```python
# most of the lags are very close to zero as expected
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x0, ax=ax);
```

```python
# create an AR(1) process
x1 = [0]
for i in range(1000):
    x = 0.5 * x1[-1] + 0.1 * np.random.randn()
    x1.append(x)
x1 = np.array(x1)
```

```python
# the AR(1) isn't obvious from this graph
plt.plot(x1);
```

```python
# redo the PACF plot and note the AR of order 1
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x1, ax=ax);
```

```python
# create another AR(1) process with -0.5 instead of 0.5
x1 = [0]
for i in range(1000):
    x = -0.5 * x1[-1] + 0.1 * np.random.randn()
    x1.append(x)
x1 = np.array(x1)
```

```python
# plot the series again
plt.plot(x1);
```

```python
# redo the PACF plot and note again the AR of order 1 and now we see the negative
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x1, ax=ax);
```

```python
# next let's do an AR(2) process
x2 = [0, 0]
for i in range(1000):
    x = 0.5 * x2[-1] - 0.3 * x2[-2] + 0.1 * np.random.randn()
    x2.append(x)
x2 = np.array(x2)
```

```python
# plot the series
plt.plot(x2);
```

```python
# plot the PACF and note the AR of order 2
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x2, ax=ax);
```

```python
# let's do an AR(5) process
x5 = [0, 0, 0, 0, 0]
for i in range(1000):
    x = 0.5 * x5[-1] - 0.3 * x5[-2] - 0.6 * x5[-5] + 0.1 * np.random.randn()
    x5.append(x)
x5 = np.array(x5)
```

```python
# plot the series - this doesn't look like just noise anymore
plt.plot(x5);
```

```python
# plot the PACF and note the AR of order 5 as expected
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x5, ax=ax);
```

```python
# look at ACF of iid noise
fig, ax = plt.subplots(figsize=(10,5))
plot_acf(np.random.randn(1000), ax=ax);
```

```python
# generate MA(1) process
errors = 0.1 * np.random.randn(1000)
ma1 = []
for i in range(1000):
    if i >= 1:
        x = 0.5 * errors[i-1] + errors[i]
    else:
        x = errors[i]
    ma1.append(x)
ma1 = np.array(ma1)
```

```python
# plot the series
plt.plot(ma1);
```

```python
# plot the ACF and note the order of 1
fig, ax = plt.subplots(figsize=(10,5))
plot_acf(ma1, ax=ax);
```

```python
# let's do an MA(2) process
errors = 0.1 * np.random.randn(1000)
ma2 = []
for i in range(1000):
    x = 0.5 * errors[i-1] - 0.3 * errors[i-2] + errors[i]
    ma2.append(x)
ma2 = np.array(ma2)
```

```python
# plot the series
plt.plot(ma2);
```

```python
# plot the ACF and note the order of 2
fig, ax = plt.subplots(figsize=(10,5))
plot_acf(ma2, ax=ax);
```

```python
# let's do an MA(3) process
errors = 0.1 * np.random.randn(1000)
ma3 = []
for i in range(1000):
    x = 0.5 * errors[i-1] - 0.3 * errors[i-2] + 0.7 * errors[i-3] + errors[i]
    ma3.append(x)
ma3 = np.array(ma3)
```

```python
# plot the series again - definitely not noise
plt.plot(ma3);
```

```python
# plot the ACF and note the order of 3
fig, ax = plt.subplots(figsize=(10,5))
plot_acf(ma2, ax=ax);
```

```python
# just to compare look at the PACF of this process we know is MA(3)
# so we know what this looks like
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(ma2, ax=ax);
```

```python
# similary look at the ACF of the process we know is AR(5)
# so we also know what this looks like
fig, ax = plt.subplots(figsize=(10,5))
plot_acf(x5, ax=ax);
```
