# Adjusted Close

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# use the csv file we created earlier
df = pd.read_csv('./data/sp500sub.csv', index_col='Date', parse_dates=True)
```

```python
# view the dataframe
df.head()
```

```python
# create a dataframe for google
goog = df[df['Name'] == 'GOOG']
```

```python
# view goog
goog.head()
```

```python
# plot the close and adjusted close prices for goog
goog[['Close', 'Adj Close']].plot(figsize=(10,10))
plt.show()
```

Note that Google had a stock split in 2014, but here our Close and Adj Close match, so our Close column already accounts for stock splits.

```python
# let's do the same thing for Apple
aapl = df[df['Name'] == 'AAPL']
aapl[['Close', 'Adj Close']].plot(figsize=(10,10))
plt.show()
```

We can see that the stock split in 2014 is already accounted for, as we saw with Google, but these differ based on dividends (Apple pays dividends while Google does not).
